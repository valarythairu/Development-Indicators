```python
KE_Series = pd.read_excel('KE_Series.xlsx')

KE_Series.head()
```

```python
Economic_Indicators = ["Expense (current LCU)"]
Economic_data= Ke_datas[Ke_datas['Indicator Name'].isin(Economic_Indicators)]
Economic_data
```

```python
Expense_data = pd.merge(Economic_data,KE_Series,right_on="Series Code",left_on="Indicator Code")
Expense_data
```

```python
Expense_data.columns
```

```python
necessary_columns = ['Country Name','Indicator Name_x','Indicator Code','year','Score','Topic','Long definition','Periodicity']

Expense_data = Expense_data[necessary_columns]

Expense_data
```

```python
Expense_data['Score'].isna().sum()
```

```python
Expense_data.isna().sum()
```

```python
Expense_data.dropna(inplace=True)
```

```python
# Convert 'year' column to datetime format
Expense_data['year'] = pd.to_datetime(Expense_data['year'])

# # Extract the year component
Expense_data['year'] = Expense_data['year'].dt.year

# # Set 'year' as the DataFrame's index
Expense_data.set_index('year', inplace=True)
```

```python
Expense_data = Expense_data.rename(columns={'Indicator Name_x':'Indicator Name'})
Expense_data
```

```python
Expense_data.info()
```

```python
# Convert Score column to numeric
Expense_data['Score'] = pd.to_numeric(Expense_data['Score'], errors='coerce')
# Drops missing values in Score Column
Expense_data.dropna(subset=['Score'], inplace=True)

# Checking th data type of Score column
Expense_data['Score'].values.dtype
```

```python
# Plotting the score for the Expense data
selected_years = Expense_data.index.unique()
scores = Expense_data['Score'].apply(lambda x: '{:.2f}'.format(x))
year = Expense_data.index

# Create a trace for the line plot
trace = go.Scatter(x=year, y=scores, mode='lines', name='Score')

# Create the layout for the plot
layout = go.Layout(
    title="Expense allocated to Kenyan Government from (2014-2021)",
    xaxis=dict(title='Year'),
    yaxis=dict(title='Score'),
    legend=dict(font=dict(size=12), orientation='h', x=0, y=1.1)
)

# Create the figure and add the trace to it
fig = go.Figure(data=[trace], layout=layout)

# Show the plot
fig.show()
```

Acording to the graph above there has been an unprise in Expense; the amount of money the Kenyan government spends on operating activities in every sector of the economy combined. This was a good sign considering the rise in taxes over the years.

```python
unique_years = Expense_data.index.unique()
adf_results = {}
threshold = 0.05  # Threshold for determining stationarity

for year in unique_years:
    # Fetch the time series data for the specific year
    year_data = Expense_data.loc[:, 'Score']

    # Check if the data is constant
    if year_data.std() == 0:
        adf_results[year] = 'Constant Data'
    else:
        # Perform the ADF test
        result = adfuller(year_data)
        p_value = result[1]

        # Determine stationarity based on the p-value
        if p_value < threshold:
            adf_results[year] = 'Stationary'
        else:
            adf_results[year] = 'Not Stationary'

# Print the stationarity results
for year, result in adf_results.items():
    print("Year:", year)
    print("Stationarity:", result)
    print()
```

Using the adullfer test and the visualization above using plotly library, it was confirmed that the time series data did not have any trends or seasonality hence it was suitable for predictive modelling without use of methods to remove non-stationarity.
