```python
KE_Series = pd.read_excel('KE_Series.xlsx')

KE_Series.head()
```

```python
Economic_Indicators = ["Expense (current LCU)"]
Economic_data= Ke_datas[Ke_datas['Indicator Name'].isin(Economic_Indicators)]
Economic_data
```

```python
Expense_data = pd.merge(Economic_data,KE_Series,right_on="Series Code",left_on="Indicator Code")
Expense_data
```

```python
Expense_data.columns
```

```python
columns = ['Country Name','Indicator Name_x','Indicator Code','year','Score','Topic','Long definition','Periodicity']

Expense_data = Expense_data[columns]

Expense_data
```

```python
Expense_data['Score'].isna().sum()
```

```python
Expense_data.isna().sum()
```

```python
Expense_data.dropna(inplace=True)
```

```python
Expense_data.loc[:,'year'] = pd.to_datetime(Expense_data['year'], format='%Y')
```

```python
Expense_data.set_index('year',inplace=True)
```

```python
Expense_data = Expense_data.rename(columns={'Indicator Name_x':'Indicator Name'})
Expense_data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.lineplot(x=Expense_data.index,y=Expense_data['Score'])
plt.xlabel('year')
plt.ylabel('Score')
plt.show()
```
